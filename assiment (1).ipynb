# *patient class*

```python
import os

class Patient:

    def __init__(self, name: str, age: int, gender: str, disease: str, patient_id: int, visit_history: str):
        self.name = name
        self.age = age
        self.gender = gender
        self.disease = disease
        self.patient_id = patient_id
        self.visit_history = visit_history

    def show_patient_info(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Disease: {self.disease}")
        print(f"Patient ID: {self.patient_id}")
        print(f"Visit History: {self.visit_history}")

    def is_duplicate(self, filename="patients.txt"):
        if not os.path.exists(filename):
            return False

        with open(filename, "r") as file:
            content = file.read()
            if f"Patient ID: {self.patient_id}" in content:
                return True
        return False

    def save_to_file(self, filename="patients.txt"):
        if self.is_duplicate(filename):
            print("Duplicate patient ID found. Record not saved.")
            return

        with open(filename, "a") as file:
            file.write("----- Patient Record -----\n")
            file.write(f"Name: {self.name}\n")
            file.write(f"Age: {self.age}\n")
            file.write(f"Gender: {self.gender}\n")
            file.write(f"Disease: {self.disease}\n")
            file.write(f"Patient ID: {self.patient_id}\n")
            file.write(f"Visit History: {self.visit_history}\n")
            file.write("\n")

        print("Patient record saved successfully.")
```

# doctor class

```python
class Doctor:

    def __init__(self, doctor_id: int, name: str, age: int,
                 gender: str, specialty: str, availability):

        self.doctor_id = doctor_id
        self.name = name
        self.age = age
        self.gender = gender
        self.specialty = specialty
        self.availability = availability if availability else []

    def show_doctor_info(self):
        print("----- Doctor Information -----")
        print(f"Doctor ID: {self.doctor_id}")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Gender: {self.gender}")
        print(f"Specialty: {self.specialty}")
        print(f"Booked Slots: {self.availability}")

    def is_available(self, time_slot: str):
        return time_slot not in self.availability

    def book_appointment(self, time_slot: str):
        if self.is_available(time_slot):
            self.availability.append(time_slot)
            print("Appointment booked successfully.")
        else:
            print("Doctor is not available at this time.")

    def cancel_appointment(self, time_slot: str):

        if time_slot in self.availability:
            self.availability.remove(time_slot)
            print("Appointment cancelled.")
        else
            print("No appointment found at this time.")
```

# Appointment class

```python
from datetime import datetime


class Appointment:

    def __init__(self, appointment_id: int, patient_id: int,
                 doctor_id: int, appointment_date: str):

        self.appointment_id = appointment_id
        self.patient_id = patient_id
        self.doctor_id = doctor_id
        self.appointment_date = appointment_date

    def validate_date(self):
        try:
            appointment_dt = datetime.strptime(self.appointment_date, "%Y-%m-%d %H:%M")
            if appointment_dt < datetime.now():
                raise ValueError("Appointment date cannot be in the past.")
            return True
        except ValueError as e:
            print(f"Invalid date: {e}")
            return False
```

# hosiptal


```python


class HospitalSystem:

    def __init__(self):
        self.patients = {}
        self.doctors = {}
        self.appointments = {}

    def add_patient(self, patient):
        if patient.patient_id in self.patients:
            print("Patient already exists.")
            return
        self.patients[patient.patient_id] = patient


    def add_doctor(self, doctor):
        if doctor.doctor_id in self.doctors:
            print("Doctor already exists.")
            return
        self.doctors[doctor.doctor_id] = doctor



    def create_appointment(self, appointment):
        try:
            if appointment.appointment_id in self.appointments:
                raise ValueError("Appointment ID already exists.")

            if appointment.patient_id not in self.patients:
                raise ValueError("Patient does not exist.")

            if appointment.doctor_id not in self.doctors:
                raise ValueError("Doctor does not exist.")

            if not appointment.validate_date():
                return

            doctor = self.doctors[appointment.doctor_id]


            if appointment.appointment_date in doctor.availability:
                raise ValueError("Doctor is not available at this time.")

            doctor.book_appointment(appointment.appointment_date)

            self.appointments[appointment.appointment_id] = appointment
            print("Appointment created successfully.")

        except Exception as e:
            print(f"Error: {e}")



    def save_data(self):
        try:
            with open("patients.txt", "w") as f:
                for p in self.patients.values():
                    f.write(f"{p.patient_id},{p.name},{p.age},{p.gender},{p.disease},{p.visit_history}\n")

            with open("doctors.txt", "w") as f:
                for d in self.doctors.values():
                    f.write(f"{d.doctor_id},{d.name},{d.age},{d.gender},{d.specialty}\n")

            with open("appointments.txt", "w") as f:
                for a in self.appointments.values():
                    f.write(f"{a.appointment_id},{a.patient_id},{a.doctor_id},{a.appointment_date}\n")

            print("Data saved successfully.")

        except Exception as e:
            print(f"Error saving data: {e}")
```

# menu

```python
def show_menu():
    print("\nHospital System Menu")
    print("1. Add Patient")
    print("2. Add Doctor")
    print("3. Create Appointment")
    print("4. Show Patients")
    print("5. Show Doctors")
    print("6. Save Data")
    print("7. Load Data")
    print("8. Exit")


def main():
    system = HospitalSystem()

    while True:
        show_menu()
        choice = input("Enter your choice: ")

        try:
            if choice == "1":
                system.menu_add_patient()

            elif choice == "2":
                system.menu_add_doctor()

            elif choice == "3":
                system.menu_create_appointment()

            elif choice == "4":
                system.menu_show_patients()

            elif choice == "5":
                system.menu_show_doctors()

            elif choice == "6":
                system.save_data()

            elif choice == "7":
                system.load_data()

            elif choice == "8":
                print("Exiting system...")
                break

            else:
                print("Invalid choice. Please select a valid option.")

        except Exception as e:
            print(f"Error: {e}")

    main()
```
